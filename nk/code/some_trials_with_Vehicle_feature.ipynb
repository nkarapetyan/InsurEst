```python
def scatter_plot(data, x, y):
    f = plt.figure()
    ax = f.add_subplot(1, 1, 1)
    ax.scatter(data[x], data[y])
    plt.title(x + ' and ' + y + ' distribution')
    plt.xlabel(x)
    plt.ylabel(y)
    #plt.show()
    return f
```
